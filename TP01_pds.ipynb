```
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import image
```

```
#--------- CONVOLUTION ---------#

# Q1

def conv2d(M, img): 
  M = np.flip(M) # flips twice
  img_w, img_h = img.shape
  M_w, M_h = M.shape
  padding = M_w//2 # padding generates edges to allow multiplication
  img_padded = np.pad(img, padding)
  img_pad_w, img_pad_h = img_padded.shape
  img_res_w = img_pad_w-M_w # resulting image goes back to original size (no padding)
  img_res_h = img_pad_h-M_h
  img_res = np.zeros((img_res_w, img_res_h)).astype(np.float32)

  for x in range (img_res_w):
    for y in range (img_res_h):
      img_res[x,y] = np.sum(M * img_padded[x:x+M_w, y:y+M_h])
  return img_res
```

```
# Q2

M = np.array([[13, 25, 12],  # M  Y  L
              [5,  14, 1 ],  # E  N  A
              [7,  1,  2 ]]) # G  A  B

img = image.imread('barbara.png')

img_final = conv2d(M, img) 

plt.figure(figsize=(8,8))
plt.imshow(img_final, cmap = plt.get_cmap('gray'))
```
